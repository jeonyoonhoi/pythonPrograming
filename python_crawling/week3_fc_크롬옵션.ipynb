## Headless Chrome
### headless 란?
- request 는 렌더링 작업이 없기 때문에 selenium보다 빠름
- headless 모드라는건~~~ 렌더링 없는 selenium 이라고 이해하자

- Headless = '창이 없는'  
- 크롬 브라우저를 이용해 인터넷 브라우징 할 때!
    - 창이 뜨고 
    - HTML 파일을 불러오고
    - CSS 파일을 불러와
    - 어떤 내용을 화면에 그려야 할지 계산
    -  ==> 이 작업을 브라우저가 자동으로 진행해준다. 
- 이런 방식을 사용할 경우 OS 따라 실행 여부가 달라짐
- 에를 들어 우분투 서버와 같은 OS 에는 화면 자체 (GUI) 가 존재하지 않음
- headless모드는 브라우저 창을 GUI , 운영체제의 창으로 띄우지 않음
- 대신 화면을 그려주는 작업인 렌더링을 가상으로 진행함
- ==> 실제 브라우저와 동일하지만 창은 뜨지 않는 방식으로 진행

### 왜 크롬을 사용하는가?
- PhantomJS(팬텀) 브라우저
    - 더 이상 개발을 진행하지 않음.
- 우리가 크롤러를 만드는 상황이 대부분 크롬에서 진행
- 결과물 그대로 가져오기 위해서는 브라우저로 크롬을 사용하는 것이 좋음
- But, 팬텀은 WebDriver Binary만으로 추가적인 설치 없이 환경을 만들 수 있다는 장점이 있음


### 어떻게 사용하는지!?!!

```python

#headless 옵션을 설정 해 준다. 
opt.add_argument('headless')
driver = webdriver.Chrome('C:/Users/YOONHOI/Desktop/chromedriver.exe',chrome_options = opt)
```

```python
driver.get("https://beomi.github.io/2017/09/28/HowToMakeWebCrawler-Headless-Chrome/")
print(driver.page_source)
```

#### 이용 2
- 쇼핑몰이나 기타 해상도에 따라 화면 배치가 달라지는 경우!
- 그럼 제일많이 나오는 걸로 바꿔주면 항상 일정한 해상도로 나오겠죠?

```python
#생성자로 새로 opt2 인스턴스를 만들어준다 
opt2 = webdriver.ChromeOptions()

opt2.add_argument('window_size=1920x1080')

# headless 모드 적용 안한 드라이버 하나 더 생성
driver_2 = webdriver.Chrome('C:/Users/YOONHOI/Desktop/chromedriver.exe',chrome_options = opt2)
```

```python
driver_2.get("https://beomi.github.io/2017/09/28/HowToMakeWebCrawler-Headless-Chrome/")
```
